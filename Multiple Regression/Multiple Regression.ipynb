```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../DataSets/Boston Housing/housing.data', delimiter=r"\s+")
column_names= ['Crime', 'Residential', 'Non', 'Charles', 'Nitric',
              'Rooms', 'Owner', 'Weighted', 'Radial',
              'Property', 'Pupil', 'Blacks', 'Lower', 'Median']

df.shape
df.columns = column_names
```

```python
df.head()
```

```python
x = df[column_names[:-1]]
y = df[column_names[-1]]
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
X_constant = sm.add_constant(x)
X_constant
```

```python
model = sm.OLS(y,X_constant)
```

```python
lr = model.fit()
```

```python
lr.summary()
```

```python
model_formula = smf.ols(formula = ' Median ~ Crime + Residential + Non + Charles + Nitric + Rooms + Owner + Weighted + Radial + Property + Pupil + Blacks + Lower' ,data =df )
fitted_formula = model_formula.fit()
```

```python
fitted_formula.summary()
```
