```python
import numpy as np
import sklearn
import matplotlib.pyplot as plt
```

```python
x = 10 * np.random.rand(100,2)
x
```

```python
slope =  np.random.rand(2) * 10 
slope
```

```python
y = np.matmul(slope,x.T) + np.random.rand(100) * 12
y
```

```python
plt.scatter(range(100),y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model2D = LinearRegression(fit_intercept= True)
```

```python
model2D
```

```python
X = x.reshape(-1,2)
X.shape
```

```python
y.shape
```

```python
model2D.fit(X,y)
```

```python
model2D.coef_
```

```python
model2D.intercept_
```

```python
x_fit = np.linspace(-1,11,100)
x_fit.shape
```

```python
X_fit = x_fit.reshape(-1,1)
X_fit.shape
```

```python
model = LinearRegression(fit_intercept= True)
```

```python
model.fit(X_fit,y)
```

```python
y_fit = model.predict(X_fit)
```

```python
plt.scatter(x_fit,y)
plt.plot(X_fit,y_fit)
```

```python
x = 10 * np.random.rand(100,3)
x
slope =  np.random.rand(3) * 10 
slope

y = np.matmul(slope,x.T) + np.random.rand(100) * 12
y
```

```python
model = LinearRegression(fit_intercept= True)
model.fit(x,y)

model.coef_
```

```python
y_predict = model.predict(x)
y_predict

plt.scatter(range(100),y-y_predict)
```
